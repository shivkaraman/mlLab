# Visualize n-dimensional data using 3D surface plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [5, 7, 10, 12]
```

#The np.meshgrid(x, y) function is used to create a grid of coordinates based on the values in the x and y lists. It returns two 2-dimensional arrays, X and Y, representing the x and y coordinates of each point in the grid.

```python
X, Y = np.meshgrid(x, y)
X, Y
```

```python
def z_function(x, y):
    return np.sin(np.sqrt(x**2+y**2))
```

```python
Z = z_function(X,Y)
Z
```

```python
plt.figure(figsize=(5,5))
ax=plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
```
